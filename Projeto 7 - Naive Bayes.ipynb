# 1) Comece carregando as bases de treino (X e y) e teste (X e y).
Verifique se o número de linhas condiz, se as variáveis estão corretas sendo apenas a de score para y e as demais nas bases de X e por último, se Y está balanceada no teste.

### 1) Separar X e y
variável-alvo(target) será : Score_Credito

```python
# Separar variáveis preditoras e a variável alvo
X = df.drop('Score_Credito', axis=1)
y = df['Score_Credito']
```

# 2) Aplique o algoritmo de Naive Bayes aos dados de treinamento.

### Aqui dividirei os dados em treino e teste e treinamos o modelo Gaussian Naive Bayes com os dados de treino.


```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

# Separar em treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Instanciar e treinar o modelo
modelo_nb = GaussianNB()
modelo_nb.fit(X_train, y_train)
```

# 3) Faça a avaliação do modelo com os dados de treinamento.
Traga a acurácia, recall e plote a matriz de confusão. Não se esqueça de avaliar com suas palavras o desempenho do modelo, interpretando as métricas.

### Avaliarei o modelo com o conjunto de treino utilizando métricas como acurácia, recall e matriz de confusão.

Dica: Para calcularmos o recall em classificação multi classe precisamos usar o atributo macro:
recall = recall_score(y_train, y_pred_train, average='macro')

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Previsão com base de treino
y_pred_train = modelo_nb.predict(X_train)

# Avaliação
print("Matriz de Confusão (Treino):")
print(confusion_matrix(y_train, y_pred_train))

print("\nRelatório de Classificação (Treino):")
print(classification_report(y_train, y_pred_train))

print(f"Acurácia (Treino): {accuracy_score(y_train, y_pred_train):.2f}")#seu código aqui
```

### Interpretação 
O modelo acertou 98% das amostras de treino.

As classes estão bem balanceadas e o modelo teve bom desempenho em todas elas.

O recall da classe 1 foi 100%, ou seja, todos os exemplos dessa classe foram corretamente reconhecidos.

Pouquíssimos erros (confusões) entre as classes.

# 4) Aplique o modelo aos dados de teste e realize a avaliação dos resultados, da mesma forma que fez acima. Não se esqueça de avaliar com as suas palavras e comparar o desempenho da base treino com a teste.

### Agora vou aplicar o modelo treinado aos dados de teste.  
Irei comparar os resultados com os obtidos no treino, observando possíveis diferenças entre as métricas (como overfitting ou perda de desempenho).  
Analisarei novamente a matriz de confusão, precisão, recall e f1-score.

```python
# Previsão no conjunto de teste
y_pred_test = modelo_nb.predict(X_test)

# Avaliação
print("Matriz de Confusão (Teste):")
print(confusion_matrix(y_test, y_pred_test))

print("\nRelatório de Classificação (Teste):")
print(classification_report(y_test, y_pred_test))

print(f"Acurácia (Teste): {accuracy_score(y_test, y_pred_test):.2f}")
```
